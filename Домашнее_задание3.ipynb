```python
# Для удобства загрузим модели на локальный диск:
from transformers import AutoModelForSequenceClassification, AutoTokenizer
# модель: rubert-tiny2-russian-sentiment
model_name = "seara/rubert-tiny2-russian-sentiment"
# Загружаем модель и токенизатор
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
# Сохраняем модель и токенизатор локально
model.save_pretrained("./rubert-tiny2-sentiment")
tokenizer.save_pretrained("./rubert-tiny2-sentiment")
# модель: rubert-base-cased-russian-sentiment
model_name = "seara/rubert-base-cased-russian-sentiment"
# Загружаем модель и токенизатор
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
# Сохраняем модель и токенизатор локально
model.save_pretrained("./rubert-base-cased-russian-sentiment")
tokenizer.save_pretrained("./rubert-base-cased-russian-sentiment")
```

```python
from transformers import pipeline
```

```python
# Загружаем модель и токенизатор из локальной папки, для удобства делаем это через функцию
def f_model(token_name, path):
  sentiment_pipeline = pipeline(token_name, model=path)
  return sentiment_pipeline
```

```python
## Поэкспериментируем с моделью rubert-tiny2-russian-sentiment
classifier = f_model("sentiment-analysis", "./rubert-tiny2-sentiment")
```

```python
classifier( ["Отличный товар",
            "Мне не понравилось",
            "Прекрасно выглядит, ничего не меняйте!",
            "Ничего не понял",
            "У меня не работала розетка, в остальном все отлично",] )
```

```python
## Поэкспериментируем с моделью rubert-tiny2-russian-sentiment
classifier = f_model("sentiment-analysis", "./rubert-base-cased-russian-sentiment")
```

```python
classifier( ["Отличный товар",
            "Мне не понравилось",
            "Прекрасно выглядит, ничего не меняйте!",
            "Ничего не понял",
            "У меня не работала розетка, в остальном все отлично"] )
```

### Как видим мождели предобучены несколько по разному. Если в rubert-tiny2-sentiment на фразу "Ничего не понял", реакция 'negative'с вероятностью 0.524005651473999, то в rubert-base-cased-russian-sentiment это neutral с вероятностью 0.6664650440216064. Если все же здесь речь идет об отзывах о товарах, то наверное обе модели потребуется доучить. В текущем контексте фразу "Ничего не понял" все же следует рассматривать как нейтральную. То же самое касаемо фразы "У меня не работала розетка, в остальном все отлично". Все же доля негатитва, из-за того, что не работала розетка привнесено. Но и та и другая модель определили эту фразу как positive. Толко  первая с вероятностью 0.6734204888343811, а другая 0.939444363117218. Все же у первой доля сомнения больше. Продолжим экспериментировать. По идее фраза "У меня все не работает" должна интерпретироваться как negative. Посмотрим, как справяться с этой фразом модели.

```python
classifier = f_model("sentiment-analysis", "./rubert-tiny2-sentiment")
classifier( ["У меня все не работает"])
```
